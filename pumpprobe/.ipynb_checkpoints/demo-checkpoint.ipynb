```python
import numpy as np
import pandas as pd
import scipy as sci

from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from matplotlib.pyplot import MultipleLocator
```

```python
from pump_probe.filesort import file_sort
```

```python
f_path = 'data/power/'
power_from = 0 # 提取文件名中power的部分
power_to = -12

file_info = {
    "f_path":f_path,
    "power_from": power_from,
    "power_to": power_to,
    "reverse": True,
}
```

```python
files = file_sort(file_info)
```

```python
from pump_probe.filedata import FileData as fileData
```

```python
fileData_list = []
for file in files:
    file_name = file_info["f_path"] + file
    f = fileData(file, file_info)
    f.shift_to_zero()
    fileData_list.append(f)
```

```python
n_files = len(files)
color_bar_style = "rainbow_r"
color_list = generating_color(n_files, color_bar_style)
```

```python
n_files = len(files)

plt.figure(figsize=(16,12))
plt.rcParams['xtick.direction'] = 'in' 
plt.rcParams['ytick.direction'] = 'in'

shift = 0.05

i = 0
for fileData in fileData_list:
    position = fileData.position
    time = fileData.time
    value_t = fileData.value
    power = fileData.power
    value_t = 20*value_t
    
    if i <3:
        value_t = value_t*0.3
        plt.text(time[-1]-2,value_t[-1]+0.2,r'$\times$0.3',fontdict={'size':'28','color':color_list[i]})
    if i == len(files)-1:
        plt.plot(time,value_t+0*(len(files)-i-1),label=str(int(power*10))+r" $\mu$J cm$^{-2}$",color=color_list[i],linewidth=3)
    else:
        plt.plot(time,value_t+0*(len(files)-i-1),label=str(int(power*10)),color=color_list[i],linewidth=3)
    i += 1

plt.xticks(fontsize=40)
plt.yticks(fontsize=40)
plt.xlabel('Time Delay (ps)',fontsize=40)
plt.ylabel(r'$\Delta$R/R (10$^{-4}$)',fontsize=40)
plt.legend(fontsize=17,frameon=False,loc='lower left')

set_lw = 3
ax=plt.gca()
ax.spines['bottom'].set_linewidth(set_lw)
ax.spines['left'].set_linewidth(set_lw)
ax.spines['right'].set_linewidth(set_lw)
ax.spines['top'].set_linewidth(set_lw)


for tickline in ax.xaxis.get_ticklines():
    tickline.set_markersize(10)
    tickline.set_markeredgewidth(5)
for tickline in ax.yaxis.get_ticklines():
    tickline.set_markersize(10)
    tickline.set_markeredgewidth(5)
ax.yaxis.set_ticks_position("left")
ax.xaxis.set_ticks_position("bottom")

plt.savefig('output/Power dependent time domain.eps',bbox_inches = 'tight')
plt.show()
```

```python
def double_exp(x, b, c, p, q):
    x = np.array(x)
    return b*np.exp(p*x) + c*np.exp(q*x)
```

```python
Time = np.array([])
Value_t = np.array([])
temp = np.zeros(len(files))
E_THG = np.zeros(len(files))
A_1 = np.zeros(len(files))
A_2 = np.zeros(len(files))
t_1 = np.zeros(len(files))
t_1_n = np.zeros(len(files))
t_2 = np.zeros(len(files))
t_2_n = np.zeros(len(files))
# bg
# file_bg = os.listdir('data/bg/')
# data_bg = np.loadtxt('data/bg/'+file_bg[0])
# value_t_bg = data_bg[:,2]
value_t_bg = 0
plt.figure(figsize=(12,24))
i = 0


shift = 1

b_init = 1
c_init = 1
p_init = -1/10
q_init = -1/1

for f in files:

    data = np.loadtxt(f_path+f)
    position = data[:,1]
    time = data[:,0]
#         time = data[0:-cut_time,0]

    value_t = data[:,2] - data[0,2]
    time_cutoff = -0.05
    value_t_cutoff=value_t[time>time_cutoff]
    time_ctf=time[time>time_cutoff]
#         time_ctf = time_ctf -time_ctf[0]
#         value_t = value_t[::-1]
    deg = f[power_from:power_to]
    temp_d = f[power_to+1:power_to+2]
    deg = int(deg)+int(temp_d)/10
#         (fre, value_f) = fourier_transfer(time, value_t)
#         if temp == 13.3:
#             x_13d3 = time
#             y_13d3 = value_t
#             i_13d3 = i
    popt, pcov = curve_fit(double_exp, time_ctf, value_t_cutoff, [b_init, c_init, p_init, q_init])
    b = popt[0]
    c = popt[1]
    p = popt[2]
    q = popt[3]
#         b_init = popt[0]
#         c_init = popt[1]
#         p_init = popt[2]
#         q_init = popt[3]
#         A = popt[4]
    value_fit = double_exp(time_ctf, b, c, p, q)
    A_1[i] = b
    A_2[i] = c
    t_1[i] = -p
    t_1_n[i] = pcov[1,1]
    t_2[i] = -q
    t_2_n[i] = pcov[3,3]
    temp[i] = deg

#         if deg >85:
    plt.plot(time,value_t+shift*(len(files)-i),label=str(deg)+' K',color=color_list[i])
#         plt.plot(time_ctf,value_t_cutoff+shift*(len(files)-i),color='k')
    plt.plot(time_ctf,value_fit+shift*(len(files)-i),color='k')
    plt.text(-2,(shift)*(len(files)-i)+0,str(10*deg),fontdict={'size':'28','color':color_list[i]})
    i += 1
        

# plt.title("Temperature dependent",fontsize=20)
plt.xticks(fontsize=28)
plt.yticks((),fontsize=40)
plt.xlabel('t/ps',fontsize=40)
# plt.ylabel('E/mV',fontsize=40)
set_lw = 3
ax=plt.gca();#获得坐标轴的句柄
ax.spines['bottom'].set_linewidth(set_lw);###设置底部坐标轴的粗细
ax.spines['left'].set_linewidth(set_lw);####设置左边坐标轴的粗细
ax.spines['right'].set_linewidth(set_lw);###设置右边坐标轴的粗细
ax.spines['top'].set_linewidth(set_lw);###设置右边坐标轴的粗细
# plt.xlim(0,15)
# plt.legend(fontsize=18,frameon=True,loc='lower right')
plt.savefig('output/Temperature dependent time domain.jpg')
plt.show()
```
